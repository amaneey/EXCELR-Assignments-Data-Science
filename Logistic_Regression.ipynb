```python
data0=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome','month'])
data0
```

```python
data0['default']=np.where(data0['default'].str.contains("yes"),1,0)
data0['housing']=np.where(data0['housing'].str.contains("yes"),1,0)
data0['loan']=np.where(data0['loan'].str.contains("yes"),1,0)
data0['y']=np.where(data0['y'].str.contains("yes"),1,0)
data0
```

```python
x=pd.concat([data0.iloc[:,0:10],data0.iloc[:,11:]],axis=1)
y=data0.iloc[:,10]
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
y_pred=classifier.predict(x)
y_pred
```

```python
y_pred_df=pd.DataFrame({'actual':y,'predicted_prob':y_pred})
y_pred_df
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(y,y_pred)
confusion_matrix
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
```

```python
fpr,tpr,thresholds=roc_curve(y,classifier.predict_proba(x)[:,1])
plt.plot(fpr,tpr,color='red')
auc=roc_auc_score(y,y_pred)

plt.plot(fpr,tpr,color='red',label='logit model')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('FalsePositive Rate or [1-True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()
```
